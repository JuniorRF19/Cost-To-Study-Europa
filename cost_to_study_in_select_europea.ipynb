Data by: https://data.world/create-a-project?sourceid=2024w5-the-cheapest-countries-to-study-in-europe-in-2023&sourcens=makeovermonday

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filename = '/content/drive/MyDrive/CodingDojo/Bootcamp - Data Science/Cores/DashBoards/cost_to_study_in_select_europea.csv'
df = pd.read_csv(filename)
df.head()
```

```python
df.info()
```

```python
df_total_living_cost = df.sort_values(by = 'total_yearly_living_costs_and_fees_ps' , ascending=False)
```

```python
df_total_living_cost['total_monthly_living_costs_and_fees_ps'] = round(df_total_living_cost['total_yearly_living_costs_and_fees_ps']/12,2)
```

```python
df_total_living_cost
```

```python
plt.figure(figsize=(16, 8))
ax = sns.barplot(df_total_living_cost, x="total_monthly_living_costs_and_fees_ps", y="country",hue="country", legend=False)

for container in ax.containers:
    ax.bar_label(container, fontsize=10)

plt.title('Costo Total Mensual de Vida por Pas')
plt.xlabel('Costos Totales Mensuales')
plt.ylabel('Pas')
#plt.legend(loc='lower right')
plt.show()
```

```python
df_total_tuition_cost = df_total_living_cost.sort_values(by = 'average_yearly_tuition_fees_ps',ascending = False)
plt.figure(figsize=(16, 8))
ax = sns.barplot(df_total_tuition_cost, x="average_yearly_tuition_fees_ps", y="country",hue="country", legend=False)

for container in ax.containers:
    ax.bar_label(container , fontsize=10)

plt.title('Costo Anual de Matricula por Pas')
plt.xlabel('Costos de Matricula Anuales')
plt.ylabel('Pas')
#plt.legend(loc='lower right')
plt.show()
```
